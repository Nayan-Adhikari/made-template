```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_population = pd.read_csv('../../data/population_USA.csv')
df_population
```

```python
df_unemployed = pd.read_excel('../../data/unemployed_reates_USA.xlsx')
df_unemployed
```

```python
df_population.isnull().sum()                #No of null values
```

```python
df_unemployed.isnull().sum()                                        #No of null values
```

```python
df_unemployed.drop([0,1,2,3,4],inplace = True)        #drop rows
df_unemployed
```

```python
df_unemployed.reset_index(drop=True,inplace = True)         # reset the index
df_unemployed
```

```python
df_unemployed = df_unemployed.dropna(axis = 1,how = 'all')                     # drop all columns with missing values
df_unemployed
```

```python
df_unemployed = df_unemployed.dropna(axis = 0,how = 'all')                         # drop all rows with missing values
df_unemployed
```

```python
df_unemployed.columns = df_unemployed.iloc[0]                               # set the row value as their column name
df_unemployed = df_unemployed[1:].reset_index(drop = True)
df_unemployed
```

```python
df_unemployed.columns.value_counts()               # count the ambiguous column names
```

```python
new_column_names = ['State']
years = list(range(2023,1999,-1))

for year in years:
    new_column_names.append(f"Rate_{year}")                 #Rename the column Rate_<YEAR>, Rank_<YEAR>
    new_column_names.append(f"Rank_{year}")

df_unemployed.columns = new_column_names
df_unemployed
```

```python
df_unemployed.columns.nunique()
```

```python
numeric_columns = df_unemployed.columns[1:]                 # change the datatype from object to float
df_unemployed[numeric_columns] = df_unemployed[numeric_columns].apply(pd.to_numeric, errors = 'coerce')
df_unemployed
```

```python
df_unemployed = df_unemployed.fillna(0)             # fill the missing values with 0
```
